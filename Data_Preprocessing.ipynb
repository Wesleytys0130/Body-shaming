## 平衡資料

```python
df = pd.read_csv("./data/all.csv",index_col=[0],low_memory=False) 
# df = pd.read_csv("./data/random_df.csv",index_col=[0],low_memory=False)
df['label'] = df['label'].astype(str)
df = df[df["label"].str.contains(pat = "0|1|2", regex=True,na=False)]
df
```

```python
df['label'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# plt.figure( figsize=(10,5) )
df['label'].value_counts().plot( kind='pie',colors = sns.color_palette('pastel'), autopct='%1.2f%%',explode = [0.01,0.01,0.01] )
plt.title( 'label' )  # 圖標題
plt.ylabel( '' )
plt.show()
```
