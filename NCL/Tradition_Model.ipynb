## 特徵抽取

```python
# import warnings
# warnings.filterwarnings("ignore")
import pandas as pd
df = pd.read_csv("./data/balance.csv",index_col=[0],low_memory=False)
df
```

```python
import re

appearance_object = []
for comment in df['content']:
    comment = str(comment)
    found = re.findall('水桶|竹竿|紙片|甜不辣|難民|骨|骷髏|塑膠|火腿|騎士|瓦斯', comment)
    if len(found) > 0:
        appearance_object.append(1)
    else:
        appearance_object.append(0)
df['appearance_object'] = appearance_object

appearance_animal = []
for comment in df['content']:
    comment = str(comment)
    found = re.findall('豬|龍|猴|狗|鳥|蟲|象', comment)
    count = re.findall('歐米龍|萊豬|印象|不死鳥', comment)
    if len(found) > 0 and len(count) == 0:
        appearance_animal.append(1)
    else:
        appearance_animal.append(0)
df['appearance_animal'] = appearance_animal

body = []
for comment in df['content']:
    comment = str(comment)
    found = re.findall('臉|腿|腳|手|肚|屁|腰', comment)
    count = re.findall('打腫臉|柯屁|腰帶', comment)
    if len(found) > 0 and len(count) == 0:
        body.append(1)
    else:
       body.append(0)
df['body'] = body

adjective = []
for comment in df['content']:
    comment = str(comment)
    found = re.findall('醜|窮|胖|肥|矮|短|粗|臭|噁|垂|厚|禿|油|老|扁|髒|魯|殘|肉|腫|低能|不正|假掰|心機|嘔', comment)
    count = re.findall('安迪魯伊茲|胖虎|胖卡|金小胖|矮油|矮堅', comment)
    if len(found) > 0 and len(count) == 0:
        adjective.append(1)
    else:
        adjective.append(0)
df['adjective'] = adjective

shame = []
for comment in df['content']:
    comment = str(comment)
    found = re.findall('垃圾|廢物|馬的|媽的|操|幹|幹你媽|幹你娘|死|要命|跩|賤', comment)
    count = re.findall('幹嘛|幹什麼|幹麼', comment)
    if len(found)>0 and len(count)==0:
        shame.append(1)
    else:
        shame.append(0)
df['shame'] = shame

df['label'] = df['label'].astype(int)
df['appearance_object'] = df['appearance_object'].astype(int)
df['appearance_animal'] = df['appearance_animal'].astype(int)
df['body'] = df['body'].astype(int)
df['adjective'] = df['adjective'].astype(int)
df['shame'] = df['shame'].astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder() 
features_tag = df[['tag']].apply(LabelEncoder().fit_transform)
features_tag
df.insert(2,"tag_num",features_tag,True)
```

## 把資料分成8:2的訓練集和測試集

```python
from sklearn.model_selection import train_test_split

df['label'] = df['label'].astype(str)
df = df[df["label"].str.contains(pat = "0|1|2", regex=True,na=False)]

all_features = df[['tag_num','appearance_object', 'appearance_animal','body','adjective','shame']]
all_labels = df['label']

# feat_train, feat_test, label_train, label_test = train_test_split(all_features, all_labels, train_size=0.8, random_state=50)
feat_train, feat_test, label_train, label_test = train_test_split(all_features, all_labels, train_size=0.8)
```

```python
label_train.value_counts()
```

## naive bayes model

```python
from sklearn.naive_bayes import MultinomialNB
nb_model = MultinomialNB()
nb_model.fit(feat_train, label_train) 
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report

predicted_nb = nb_model.predict(feat_test)
bayes_accuracy = accuracy_score(label_test, predicted_nb)
bayes_fscore = f1_score(label_test, predicted_nb, average='micro') 

print(bayes_accuracy)
print(bayes_fscore)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(nb_model, random_state=1).fit(feat_test, label_test)
eli5.show_weights(perm, feature_names = feat_test.columns.tolist())
```

```python
report_nb = classification_report(label_test, predicted_nb)
print(report_nb)
```

## Decision Tree

```python
from sklearn import tree

dt_model = tree.DecisionTreeClassifier()
dt_model.fit(feat_train, label_train)
```

```python
predicted_dt = dt_model.predict(feat_test)
dt_accuracy = accuracy_score(label_test, predicted_dt)
dt_fscore = f1_score(label_test, predicted_dt, average='micro') 

print(dt_accuracy)
print(dt_fscore)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(dt_model, random_state=1).fit(feat_test, label_test)
eli5.show_weights(perm, feature_names = feat_test.columns.tolist())
```

```python
report_dt = classification_report(label_test, predicted_dt)
print(report_dt)
```

## SVM

```python
from sklearn import svm

svm_model = svm.SVC(kernel='poly', C=1)
svm_model.fit(feat_train, label_train)
```

```python
predicted_svm = svm_model.predict(feat_test)
svm_accuracy = accuracy_score(label_test, predicted_svm)
svm_fscore = f1_score(label_test, predicted_svm, average='micro') 

print(svm_accuracy)
print(svm_fscore)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(svm_model, random_state=1).fit(feat_test, label_test)
eli5.show_weights(perm, feature_names = feat_test.columns.tolist())
```

```python
report_svm = classification_report(label_test, predicted_svm)
print(report_svm)
```
