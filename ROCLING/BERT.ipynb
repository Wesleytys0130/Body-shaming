```python
import pandas as pd
from datasets import Dataset
from transformers import AutoConfig
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
from transformers import DataCollatorWithPadding
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn import metrics
import matplotlib.pyplot as plt
from transformers import AutoTokenizer

pred_list = []
true_list = []


def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True)

for i in range(1,10):
    df = pd.read_csv("./data/llms_df.csv",index_col=[0],low_memory=False)
    sr = float(f"0.{i}")
    n_seq = df.shape[0]
    n_train = int(n_seq*sr)
    # title, lable
    train_data = {
        "text": df.content[:n_train],
        "label": df.label[:n_train]
    }
    test_data = {
        "text": df.content[n_train:],
        "label": df.label[n_train:]
    }
    
    tokenizer = AutoTokenizer.from_pretrained("bert-base-chinese")
    
    train_ds = Dataset.from_dict(train_data)
    test_ds = Dataset.from_dict(test_data)
    train_ds = train_ds.map(preprocess_function, batched=True)
    test_ds = test_ds.map(preprocess_function, batched=True)
    
    model_name = "bert-base-chinese"
    num_label = 3
    cache_dir = "cache/"
    model_version = "main"

    config = AutoConfig.from_pretrained(
            model_name,
            num_labels=num_label,
            finetuning_task="dcard-seq",
            cache_dir=cache_dir,
            revision=model_version,
    #         use_auth_token=None,
        )
    
    
    model_name = "bert-base-chinese"

    model = AutoModelForSequenceClassification.from_pretrained(
        model_name, 
        from_tf=False,
        config=config,
        cache_dir=cache_dir,
        revision=model_version)
    
    
    data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

    training_args = TrainingArguments(
        output_dir="BERT_body_shaming",
        use_mps_device=True,
        learning_rate=2e-5,
        num_train_epochs=5,
        per_device_train_batch_size=64,
        per_device_eval_batch_size=64,
        weight_decay=0.01,
        evaluation_strategy="epoch",
        save_strategy="epoch",
        load_best_model_at_end=True,
        push_to_hub= True,
    )




    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_ds,
        eval_dataset=test_ds,
        tokenizer=tokenizer,
        data_collator=data_collator,
    #     compute_metrics=compute_metrics,
    )

    trainer.train()
    
    ypred_test = trainer.predict(test_ds).predictions.argmax(axis=1)
    ytrue_test = test_ds["label"]
    
    
    pred_list.append(ypred_test)
    true_list.append(ytrue_test)
    
    print(f"這是第{i}筆\n\n")
    print(classification_report(ytrue_test, ypred_test))
    confusion_matrix = metrics.confusion_matrix(ytrue_test, ypred_test)
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1, 2])
    cm_display.plot()
    plt.show()
```

```python
dir_keys = ["pred_test","true_test"]
dir_values =[pred_list, true_list]
dic = dict(zip(dir_keys, dir_values))
df = pd.DataFrame(dic)
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# 創建一個示範的4x4混淆矩陣
cm = np.array([[163, 64, 150, 123],
               [205, 202, 65, 28],
               [238, 153, 188, 21],
               [0, 0, 0, 0]])

# 定義類別的標籤
classes = ["無歧視", "歧視為胖", "歧視為瘦","無法判斷"]

# 繪製混淆矩陣
fig, ax = plt.subplots()
im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)

# 設置類別標籤
ax.set(xticks=np.arange(cm.shape[1]),
       yticks=np.arange(cm.shape[0]),
       xticklabels=classes, yticklabels=classes,
       ylabel='實際標籤',
       xlabel='預測標籤')

# 在方塊中顯示數字
thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, format(cm[i, j], 'd'),
                ha="center", va="center",
                color="white" if cm[i, j] > thresh else "black")
cbar = ax.figure.colorbar(im, ax=ax, ticks=np.arange(0, cm.max()+1, 50))
# plt.show()
plt.savefig("llama.pdf", bbox_inches='tight')
```
